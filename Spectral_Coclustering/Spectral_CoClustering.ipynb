```python
import numpy as np
from sklearn.cluster import SpectralCoclustering
from sklearn.metrics import consensus_score
from sklearn.datasets import make_biclusters
from matplotlib import pyplot as plt
```

```python
boson_sampling_obj = np.load('bs_exp2_part1.npz')
boson_data = boson_sampling_obj['arr_0']

# convert the data into a bool data
# If the values in matrix are

plt.matshow(boson_data, cmap=plt.cm.Blues)
plt.title("Original dataset")

model = SpectralCoclustering(n_clusters=3)
n = model.fit(boson_data)

print(n.get_shape(1))

fit_data = boson_data[np.argsort(model.row_labels_)]
fit_data = fit_data[:, np.argsort(model.column_labels_)]

# score = consensus_score(model.biclusters_,(boson_data[np.argsort(model.row_labels_)],fit_data[:, np.argsort(model.column_labels_)]))
# print('The consensus score is ',score)

plt.matshow(fit_data, cmap=plt.cm.Blues)
plt.title("After biclustering; Rearranged to show biclusters for Boson Sampling")
```

```python
'''
Below code is for trying to perform biclustering on a dataset 
where the original bicluster is scattered and is tried to retrieve
'''

# Load the object for GBS 
gaussian_boson_sampling_obj = np.load('gbs_exp2_part1.npz')
# Get the dataset 
gaussian_data = gaussian_boson_sampling_obj['arr_0']

# convert the data into a bool data
plt.matshow(gaussian_data, cmap=plt.cm.Blues)
plt.title("Original dataset")

# Perform the Spectral CoClustering
model = SpectralCoclustering(n_clusters=3)
n = model.fit(gaussian_data)

# print(n.get_shape(1))

fit_data = gaussian_data[np.argsort(model.row_labels_)]
fit_data = fit_data[:, np.argsort(model.column_labels_)]

plt.matshow(fit_data, cmap=plt.cm.Blues)
plt.title("After biclustering; rearranged to show biclusters for the Gaussian Boson Sampling")
```
