```python
# Code for collated Heatmaps
import numpy as np
import matplotlib.pyplot as plt

# Load all data parts
filepaths = ["bs_exp1_part0.npy", "bs_exp1_part1.npy", "bs_exp1_part2.npy", 
             "bs_exp1_part3.npy", "bs_exp1_part4.npy", "bs_exp1_part5.npy"]
data_parts = [np.load(fp) for fp in filepaths]

# Create a figure with subplots
fig, axs = plt.subplots(3, 2, figsize=(16, 18))  # 3 rows, 2 columns

for i, ax in enumerate(axs.flat):
    im = ax.imshow(data_parts[i])
    ax.set_title(f'Heatmap for Bosonic Sampling Exp1, Part {i}')
    fig.colorbar(im, ax = ax)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
